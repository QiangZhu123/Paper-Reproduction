```python
import torch
from torch import nn
```

```python
class Swish(nn.Module):
    
    def __init__(self):
        super().__init__()
        self.beta = nn.Parameter(torch.randn(1))
    def forward(self,x):
        print(self.beta)
        return torch.sigmoid(self.beta*x)+x
```

```python
s = Swish()
x =torch.ones(1,3,10,10)
s.eval()
out = s(x)
out
```
